```python
import os
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
# Plot Experiment1

data = []
for filename in os.listdir('experiment1/cumu_rewards/'):
    with open('experiment1/cumu_rewards/' + filename, 'r') as f:
        data.append(f.readlines())
        
for i in range(len(data)):
    data[i] = [d.rstrip() for d in data[i]]
    data[i] = [float(d) for d in data[i]]

fig, ax = plt.subplots()
for i in range(len(data)):
    ax.plot([j for j in range(len(data[i]))], data[i])

plt.xlabel('Episodes')
plt.ylabel('Cumulative Reward')
plt.title('γ = 0.9, α = 0.01')

plt.savefig('experiment1plots.png', bbox_inches='tight')
```

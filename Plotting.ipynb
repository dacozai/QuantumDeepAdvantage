```python
import os
import matplotlib.pyplot as plt


%matplotlib inline
```

```python
# Plot Experiment1

data = []
for filename in os.listdir('experiment1/cumu_rewards/'):
    with open('experiment1/cumu_rewards/' + filename, 'r') as f:
        data.append(f.readlines())
        
for i in range(len(data)):
    data[i] = [d.rstrip() for d in data[i]]
    data[i] = [float(d) for d in data[i]]

fig, ax = plt.subplots()
for i in range(len(data)):
    ax.plot([j for j in range(len(data[i]))], data[i])

plt.xlabel('Episodes')
plt.ylabel('Cumulative Reward')
plt.title('γ = 0.9, α = 0.01')

plt.savefig('experiment1plots.png', bbox_inches='tight')
```

```python
fig, ax = plt.subplots(figsize=(10,10))

plt.plot([j for j in range(len(data1avg))], data1avg, 'o-')

plt.plot([j for j in range(len(data2avg))], data2avg, 'o-')

plt.plot([j for j in range(len(data3avg))], data3avg, 'o-')

plt.plot([j for j in range(len(data4avg))], data4avg, 'o-')

plt.plot([j for j in range(len(data5avg))], data5avg, 'o-')

plt.xlabel('x50 Episodes')
plt.ylabel('Frequency of Correct Solution')
plt.legend(['γ = 0.9', 'γ = 0.7', 'γ = 0.5', 'γ = 0.3', 'γ = 0.1'])

plt.title('Learning Rate 10e-2')

plt.savefig('experiment3plots_a0.01.png', bbox_inches='tight')
```

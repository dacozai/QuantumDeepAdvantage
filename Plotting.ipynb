```python
%matplotlib inline
import matplotlib.pyplot as plt

with open('upthendown', 'r') as f:
    data = f.readlines()

data = [d.rstrip() for d in data]
data = [float(d) for d in data]
    
plt.plot([i for i in range(len(data))], data)
plt.xlabel('Episodes')
plt.ylabel('Cumulative Reward')
plt.show()
```

```python
import os
import matplotlib.pyplot as plt


%matplotlib inline

data = []
for filename in os.listdir('experiment1/cumu_rewards/'):
    with open('experiment1/cumu_rewards/' + filename, 'r') as f:
        data.append(f.readlines())
        
for i in range(len(data)):
    data[i] = [d.rstrip() for d in data[i]]
    data[i] = [float(d) for d in data[i]]

fig, ax = plt.subplots()
for i in range(len(data)):
    ax.plot([j for j in range(len(data[i]))], data[i])

plt.xlabel('Episodes')
plt.ylabel('Cumulative Reward')
plt.show()
    
```
